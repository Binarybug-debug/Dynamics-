Here we will make inference of the Life Expectancy Data to understand Data Visualization and Measures of Central Tendency as taught by Mr. Noah Kandie

```python
import pandas as pd
import numpy as np
```

```python
#Reading the Data
df = pd.read_csv('Life Expectancy Data.csv')
```

```python
# 1. Data Understanding // 2D shape (rows, columns) [0,1] // 3D array (layer,row,columns)[0,1,2]
print(f"Number of observations (rows): {df.shape[0]}")
print(f"Number of variables (columns): {df.shape[1]}")
```

```python
#df.fillna(df.mean()).mean()*100 #this replacesReplace NaNs with column means
(df.isna().mean()*100).sort_values(ascending=False)
```

```python
#2. Categorical and Numerical
numerical_variable = df.select_dtypes(include = ['number']).columns.tolist()
print(f"Numerical Variables: {numerical_variable}")

categorical_variable = df.select_dtypes(exclude = ['number']).columns.tolist()
print(f"Categorical Variable: {categorical_variable}")
```

```python
# 3. Cleaned Data
df_cleaned = df.dropna()

print(f"Cleaned dataset: {df_cleaned.info()}")
print(f"Original dataset: {df.info()}")
```

```python
# 4. Central Tendancies
polio_deaths = df_cleaned['Polio']

mean_polio = np.mean(polio_deaths)
median_polio = np.median(polio_deaths)
max_polio = max(polio_deaths)
min_polio = min(polio_deaths)
std_polio = np.std(polio_deaths)
mode_polio = polio_deaths.mode()[0]

range_polio = max_polio - min_polio

print(f"Polio - Mean: {mean_polio}")
print(f"Polio - Median: {median_polio}")
print(f"Polio - Mode: {mode_polio}")
print(f"Polio - Standard Deviation: {std_polio}")
print(f"Polio - Min: {min_polio}")
print(f"Polio - Max: {max_polio}")
print(f"Polio - Range: {range_polio}")
```

```python
# Shape interpretation
if mean_polio > median_polio:
    print("The distribution is likely right-skewed.")
elif mean_polio < median_polio:
    print("The distribution is likely left-skewed.")
else:
    print("The distribution is likely symmetric.")

status_mode = df_cleaned['Country'].mode()[0]
print(f"Most common country: {status_mode}")
```
