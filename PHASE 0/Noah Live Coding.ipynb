```python
df.select_dtypes(include ='number')
```

```python



def stat_function(df):
  cols = df.columns

  for col in cols:
    print()
    if col in df._get_numeric_data().columns: # select the numeric cols
      mean_x = np.mean(df[col])
      median_x = np.median(df[col])
      mode_x = df[col].mode()  # unimodal, bimodal....
      std_x = np.std(df[col])
      print(f'The mean of the {col} is {round(mean_x,2)} , median is {median_x} and the deviation is {round(std_x,3)}')
      print(f'there"s {len(mode_x)} modes and they are {list(mode_x)}')
      print(mean_x)

    if col not in df._get_numeric_data().columns:
      mode_x = df[col].mode()
      print(f'the mode of the {col} is {mode_x}')


stat_function(df)
```
